# Introduction to Plotting Images

1. First we need to start by importing our libraries needed for this work.

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
%matplotlib qt5
```

```python
im = np.random.random((100, 100))
```

```python
im.dtype
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(im)
plt.axis('off')
plt.subplot(122)
plt.imshow(im[25:75, 25:75])

plt.subplots_adjust(bottom=0.1, right=0.8, top=0.9)
cax = plt.axes([0.85, 0.3, 0.02, 0.4])
# ['left', 'bottom'. 'width', 'height']
plt.colorbar(cax=cax)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.imshow(im)
ax.axis('off')
plt.show()
```

```python
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import matplotlib.font_manager as fm

fontprops = fm.FontProperties(size=30)
scalebar = AnchoredSizeBar(
        transform=ax.transData, size=10, label='5 nm',
        loc=4, frameon=False,
        color='white', size_vertical=0.7,
        label_top=False,
        fontproperties=fontprops)
```

```python
ax.add_artist(scalebar)
```
